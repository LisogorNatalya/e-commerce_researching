###### Пользователи из каких стран совершили наибольшее число успешных бронирований? Вывести топ-5 стран

```python
reservation = bookings.query ('is_canceled != 1')
```

```python
country_res = reservation.groupby (['country', 'is_canceled'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False)
```

```python
country_res.head(5)
```

###### На сколько ночей в среднем бронируют отели типа City Hotel? Resort Hotel?

```python
nights_by_hotels = bookings\
    .groupby('hotel')\
    .aggregate({'stays_total_nights': 'mean'})\
    .sort_values('stays_total_nights', ascending=False)\
    .stays_total_nights.round(2)
```

```python
nights_by_hotels
```

###### В какой месяц бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 

```python
bookings.query('hotel == "City Hotel"')\
    .groupby(['arrival_date_year', 'arrival_date_month'])\
    .is_canceled.sum().groupby(['arrival_date_year']).nlargest(1)
```

###### Не все бронирования завершились успешно , поэтому можно посчитать, сколько клиентов было потеряно в процессе - посчитать метрику Churn Rate.
Давайте посмотрим, как эта метрика связана с наличием детей у клиентов!

```python
bookings['total_kids'] = bookings.children + bookings.babies
bookings['has_kids'] = bookings['total_kids'] >= 1
```

```python
rates = bookings.groupby('has_kids')['is_canceled'].value_counts(normalize=True)
round(rates * 100, 2)
```
