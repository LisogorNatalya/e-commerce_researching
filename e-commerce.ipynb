### 5. Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц

```python
customers_orders.head()
```

```python
# Добавим столбец с месяцем-годом создания заказа для формирования наших когорт
customers_orders['month_year'] = customers_orders['order_purchase_timestamp'].dt.to_period('M')
```

```python
cohort = customers_orders[['customer_unique_id', 'order_id', 'order_purchase_timestamp','month_year']]
# и добавим для каждого пользователя колонку с месяцем и годом с формированием первого заказа
cohort['cohort'] = cohort.groupby('customer_unique_id')['order_purchase_timestamp'] \
    .transform('min').dt.to_period('M')
```

Проверим, какие у нас есть периоды

```python
cohort.month_year.value_counts().sort_index()
```

Видим, что данные за 2016 и 2018 год у нас не полные, поэтому оставим для выявления retention только 2017 год.
Так же добавим значение "период", который будем показывать нам разницу между первой покупкой и датой заказа

```python
from operator import attrgetter
cohort['period_number'] = (cohort.month_year - cohort.cohort).apply(attrgetter('n'))
```

```python
cohort = cohort \
    .query("cohort >= '2017-01' and cohort <= '2017-12'")\
    .groupby(['cohort', 'period_number'], as_index=False) \
    .agg({'customer_unique_id':'nunique'}) \
    .rename(columns={'customer_unique_id':'quantity_of_customers'})\
    .pivot(index = 'cohort', columns = 'period_number', values = 'quantity_of_customers')
```

```python
# Разделим значения в каждом периоде на размер когорты
cohort=(cohort.div(cohort[0], axis='index'))
```

```python
cohort
```

```python
retention = (cohort
            .style
            .background_gradient(cmap='Blues')
            .highlight_null('white')
            .format('{:.2%}', na_rep='-'))
retention
```

### 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.

```python
items_orders
```

##### Для RFM-анализа нам потребуются определенные данные о каждом из клиентов:

ID клиента / имя / компания и т.д. - для их идентификации.

Давность (R) как количество дней с момента последней покупки: Сколько дней назад была совершена их последняя покупка? Вычтите дату последней покупки из сегодняшнего дня, чтобы рассчитать значение давности. 1 день назад? 14 дней назад? 500 дней назад?

Частота (F) как общее количество транзакций: Сколько раз клиент совершал покупки в нашем магазине? Например, если кто-то сделал 10 заказов за определенный период времени, его частота равна 10.

Деньги (М) как общая сумма потраченных денег: Сколько $$ (или любой другой валюты расчета) потратил этот клиент? Просто суммируйте деньги от всех транзакций, чтобы получить значение M.

```python
RFM_table = items_orders[['customer_unique_id', 'order_id', 'order_status', 'order_delivered_customer_date', 'product_id', 'price']]
```

Т.к. изначально мы определили покупкой заказ со статусом delivered, то оставим в таблице только заказы с этим статусом

```python
RFM_table = RFM_table.query("order_status == 'delivered'")[['customer_unique_id', 'order_id', 'order_delivered_customer_date', 'product_id', 'price']]
```

```python
RFM_table
```

В RFM анализе обычно исследуются данные за определенный период (чаще всего 365 дней). Видим, что покупок в сентябре и октябре не такое большое количество, поэтому будет разумнее исследовать данные с августа 2017 года по август 2018 для формирования полноценной картины.

```python
RFM_table['order_delivered_customer_date'].dt.to_period('M').value_counts().sort_index(ascending=False)
```

```python
# Примем за сегодняшнее число 31 августа 2018 года.
import datetime as dt
NOW = dt.datetime(2018,8,31)
RFM_table_for_year = RFM_table\
    .query("order_delivered_customer_date >= '2017-08-31' and order_delivered_customer_date <= '2018-08-31'")
```

```python
RFM_table_for_year
```

```python
RFM_table_for_year['order_delivered_customer_date'].dt.to_period('M').value_counts().sort_index(ascending=False)
```

```python
RFM_final_table=RFM_table_for_year.groupby('customer_unique_id').agg({'order_delivered_customer_date': lambda x: (NOW - x.max()).days, # Recency
                                                'order_id': lambda x: len(x.unique()), # Frequency
                                                'price': lambda x: x.sum()})    # Monetary 

RFM_final_table['order_delivered_customer_date'] = RFM_final_table['order_delivered_customer_date'].astype(int)

RFM_final_table.rename(columns={'order_delivered_customer_date': 'recency', 
                         'order_id': 'frequency',
                         'price': 'monetary_value'}, inplace=True)
```

```python
RFM_final_table
```

```python
quantiles = RFM_final_table.quantile(q=[0.25,0.5,0.75])
quantiles
```

В нашем случае сегментация по числу покупок является не совсем корректной, т.к. в первом задании мы выяснили, что больше 90% пользователей совершают в нашем магазине только одну покупку, и делить данный показатель на 4 сегмента по квантилям нет смысла.

Разделим сегменты по frequency следующим образом: 

1 - для пользователей, которые совершили только одну покупку 

3 - для пользователей, совершивших больше 1ой покупки

```python
# Converting quantiles to a dictionary, easier to use.
quantiles = quantiles.to_dict()
##  RFM Segmentation ----
RFM_Segment = RFM_final_table.copy()
# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def R_Class(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1

# Arguments (x = value, p = recency, monetary_value, frequency, k = quartiles dict)
def M_Class(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
RFM_Segment['R_Quartile'] = RFM_Segment['recency'].apply(R_Class, args=('recency',quantiles,))
RFM_Segment['F_Segment'] = RFM_Segment['frequency'].apply(lambda x: 1 if x==1 else 3)
RFM_Segment['M_Quartile'] = RFM_Segment['monetary_value'].apply(M_Class, args=('monetary_value',quantiles,))
RFM_Segment['RFMClass'] = RFM_Segment.R_Quartile.map(str) \
                            + RFM_Segment.F_Segment.map(str) \
                            + RFM_Segment.M_Quartile.map(str)
```

```python
RFM_Segment
```

```python
# Посмотрим, как разбились наши сегменты
RFM_Segment.RFMClass.value_counts()
```

```python
# ТОП-5 наших самых потерянных пользователй - совершали одну минимальную покупку 364 дня назад
RFM_Segment[RFM_Segment['RFMClass']=='111'].sort_values('recency',ascending=False).head(5)
```

```python
# ТОП-5 наших лояльных клиентов - совершили несколько покупок на сумму выше среднего
RFM_Segment[RFM_Segment['F_Segment'] == 3 ].sort_values('monetary_value', ascending=False).head(5)
```

##### Как мы видим среди наших лояльных клиентов есть такие, которые уже давно не совершали покупок. Нужно задуматься о способах их возвращения: рассылка со спецпредложениями, бонусами, акциями
